```python
data_new = spark.read.load('pandas_new.csv',format = "csv", inferSchema=True,header = "true")
data_new.printSchema()

```

```python
data = data_new.select('Label', '90D', 'RevolvingRatio', '30-59D', '60-89D', 'Age')
data.show()
```

```python
data.printSchema()
data=data.withColumn("Label",data.Label.astype("double"))
data=data.withColumn("90D",data["90D"].astype("double"))
data=data.withColumn("30-59D",data["30-59D"].astype("double"))
data=data.withColumn("60-89D",data["60-89D"].astype("double"))
data=data.withColumn("Age",data.Age.astype("double"))
data=data.withColumn("RevolvingRatio",data.RevolvingRatio.astype("double"))
data.printSchema()
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.classification import (RandomForestClassifier, GBTClassifier, DecisionTreeClassifier)
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
import findspark
```

```python
assembler = VectorAssembler(
  inputCols=['90D',
             'RevolvingRatio',
             '30-59D',
             '60-89D',
             'Age'],
              outputCol="features")
```

```python
output = assembler.transform(data)
```

```python
from pyspark.ml.feature import StringIndexer
```

```python
indexer = StringIndexer(inputCol="Label", outputCol="LabelIndex")
output_fixed = indexer.fit(output).transform(output)
```

```python
final_data = output_fixed.select("features",'LabelIndex')
train_data,test_data = final_data.randomSplit([0.7,0.3])
```

```python
dtc = DecisionTreeClassifier(labelCol='LabelIndex',featuresCol='features')
dtc_model = dtc.fit(train_data)
```

```python
dtc_predictions = dtc_model.transform(test_data)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
my_binary_eval = BinaryClassificationEvaluator(labelCol = 'LabelIndex')
```

```python
print("DTC")
print(my_binary_eval.evaluate(dtc_predictions))
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
acc_evaluator = MulticlassClassificationEvaluator(labelCol="LabelIndex", predictionCol="prediction", metricName="accuracy")
```

```python
dtc_acc = acc_evaluator.evaluate(dtc_predictions)
```

```python
print("Here are the results!")
print('-'*40)
print('A single decision tree has an accuracy of: {0:2.2f}%'.format(dtc_acc*100))
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import (VectorAssembler,VectorIndexer,
                                OneHotEncoder,StringIndexer)
```

```python
assembler = VectorAssembler(
  inputCols=['90D',
             'RevolvingRatio',
             '30-59D',
             '60-89D',
             'Age'],
              outputCol="features")
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
```

```python
log_reg_titanic = LogisticRegression(featuresCol='features',labelCol='Label')  
```

```python
pipeline = Pipeline(stages=[
                           assembler,log_reg_titanic])
```

```python
train_titanic_data, test_titanic_data = data.randomSplit([0.7,.3])
```

```python
fit_model = pipeline.fit(train_titanic_data)
```

```python
results = fit_model.transform(test_titanic_data)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

my_eval = BinaryClassificationEvaluator(rawPredictionCol='prediction',
                                       labelCol='Label')
```

```python
results.select('Label','prediction').show()
AUC = my_eval.evaluate(results)
AUC
```
